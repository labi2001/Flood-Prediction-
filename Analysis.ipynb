First, confirming the nature of the missing values and their relationship with other columns:

```python
# Check for missing values in preciptype
missing_preciptype = df['preciptype'].isnull().sum()
print(f"Missing preciptype values: {missing_preciptype}")

# Check the pattern of missing values
missing_pattern = df[df['preciptype'].isnull() & (df['precip'] == 0) & (df['precipcover'] == 0) & (df['precipprob'] == 0)]
print(f"Pattern of missing values: {missing_pattern.shape[0]} rows")

# Check remaining missing values
remaining_missing = df[df['preciptype'].isnull() & ((df['precip'] != 0) | (df['precipcover'] != 0) | (df['precipprob'] != 0))]
print(f"Remaining missing values: {remaining_missing.shape[0]}")
```

Given that the majority of missing values (3681 out of 4017) are clearly associated with no precipitation, fill them with "none":

```python
# Fill missing preciptype values where precip, precipcover, and precipprob are all 0
df.loc[(df['preciptype'].isnull()) & (df['precip'] == 0) & (df['precipcover'] == 0) & (df['precipprob'] == 0), 'preciptype'] = 'none'
```

```python
from astral import moon
```

```python
# Function to get moon phase for a specific date
# Function to get moon phase with descriptive name for a specific date
def get_moon_phase(date):
    phase = moon.phase(date)
    if phase < 1.845:  # New Moon to First Quarter
        return 'New Moon'
    elif phase < 5.537:  # First Quarter to Full Moon
        return 'First Quarter'
    elif phase < 9.230:  # Full Moon to Last Quarter
        return 'Full Moon'
    elif phase < 12.922:  # Last Quarter to New Moon
        return 'Last Quarter'
    else:
        return 'New Moon'  # Just in case of any edge cases


# Apply get_moon_phase function to each row in DataFrame
df4['moon_phase'] = df4['datetime'].apply(get_moon_phase)
```

Tidal Influence: Moon phases influence tidal patterns due to gravitational forces. Higher tides during new and full moons (spring tides) can increase the likelihood of coastal flooding, especially in combination with storm surges or heavy rainfall.

```python
# Convert moon phase to numerical categories
df4['moon_phase_cat'] = df4['moon_phase'].map({
    'New Moon': 1,
    'First Quarter': 2,
    'Full Moon': 3,
    'Last Quarter': 4
})

```

```python
# Group by moon phase categories and calculate mean flood occurrences
grouped = df4.groupby('moon_phase_cat')['flood occurrence'].mean().reset_index()

# Map numerical categories back to moon phase names for plotting
grouped['moon_phase'] = grouped['moon_phase_cat'].map({
    1: 'New Moon',
    2: 'First Quarter',
    3: 'Full Moon',
    4: 'Last Quarter'
})

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(grouped['moon_phase'], grouped['flood occurrence'], color='skyblue')
plt.xlabel('Moon Phase')
plt.ylabel('Mean Flood Occurrences')
plt.title('Mean Flood Occurrences by Moon Phase')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```


-  analysis showing higher mean flood occurrences during "First Quarter" and "Last Quarter" moon phases suggests that other factors beyond tidal effects may dominate flood occurrences in Lagos. These phases may coincide with regional climate patterns, river levels, or local drainage issues that contribute significantly to flood events.
